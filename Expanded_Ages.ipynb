```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#gene2 = PRKAR1A, gene 3 = PPP1R1B / DARPP-32
gene1 = pd.read_csv('https://raw.githubusercontent.com/tesscstanley/Neurobio-Project/main/PPP%20new%20ages.csv')
gene2 = pd.read_csv('https://raw.githubusercontent.com/tesscstanley/Neurobio-Project/main/new%20ages%20PRKA%20data.csv')
gene3 = pd.read_csv('https://raw.githubusercontent.com/tesscstanley/Neurobio-Project/main/DARPP32.csv')
```

```python
vfc1 = gene1.loc[gene1['structure_abbreviation']=='VFC']
dfc1 = gene1.loc[gene1['structure_abbreviation']=='DFC']
mfc1 = gene1.loc[gene1['structure_abbreviation']=='MFC']
vfc2 = gene2.loc[gene2['structure_abbreviation']=='VFC']
dfc2 = gene2.loc[gene2['structure_abbreviation']=='DFC']
mfc2 = gene2.loc[gene2['structure_abbreviation']=='MFC']
vfc3 = gene3.loc[gene3['structure_abbreviation']=='VFC']
dfc3 = gene3.loc[gene3['structure_abbreviation']=='DFC']
mfc3 = gene3.loc[gene3['structure_abbreviation']=='MFC']
```

```python
vfc1_mean = vfc1.groupby(['donor_age']).mean(['expression_level'])
dfc1_mean = dfc1.groupby(['donor_age']).mean(['expression_level'])
mfc1_mean = mfc1.groupby(['donor_age']).mean(['expression_level'])
vfc2_mean = vfc2.groupby(['donor_age']).mean(['expression_level'])
dfc2_mean = dfc2.groupby(['donor_age']).mean(['expression_level'])
mfc2_mean = mfc2.groupby(['donor_age']).mean(['expression_level'])
vfc3_mean = vfc3.groupby(['donor_age']).mean(['expression_level'])
dfc3_mean = dfc3.groupby(['donor_age']).mean(['expression_level'])
mfc3_mean = mfc3.groupby(['donor_age']).mean(['expression_level'])
fig, ax = plt.subplots(3)
x = ['0.3','1.0','2.0','3.0','4.0','8.0','11.0','13.0','18.0','19.0','21.0','23.0','30.0']
x1 = ['0.3','0.8','1.0','2.0','3.0','4.0','8.0','11.0','13.0','18.0','19.0','21.0','23.0','30.0']
x2 = ['0.3','0.8','1.0','2.0','3.0','4.0','8.0','11.0','13.0','18.0','19.0','21.0','30.0']
v = vfc1_mean['expression_level']
d = dfc1_mean['expression_level']
m = mfc1_mean['expression_level']


ax[0].bar(x,v,align='center',color='lavender',label='PPP3CB Expression, VFC')
ax[0].plot(x,v,color='black')

ax[1].bar(x2,d,align='center',color='darkseagreen',label='PPP3CB Expression, DFC')
ax[1].plot(x2,d,color='black')

ax[2].bar(x,m,align='center',color='powderblue',label='PPP3CB Expression, MFC')
ax[2].plot(x,m,color='black')

ax[0].legend()
ax[1].legend()
ax[2].legend()
print(vfc3_mean)
```
