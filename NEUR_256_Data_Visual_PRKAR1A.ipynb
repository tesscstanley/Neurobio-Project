```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#PPP3CB = gene1, PRKAR1A = gene2
#upload both gene files as pandas dataframes
gene1 = pd.read_csv('https://raw.githubusercontent.com/abharr/Neurobio-Project/main/PPP3CB%20Data.csv')
gene2 = pd.read_csv('https://raw.githubusercontent.com/abharr/Neurobio-Project/main/PRKAR1A%20Data.csv')
```

```python
#separate data into brain region
vfc1 = gene1.loc[gene1['structure_abbreviation']=='VFC']
dfc1 = gene1.loc[gene1['structure_abbreviation']=='DFC']
mfc1 = gene1.loc[gene1['structure_abbreviation']=='MFC']
vfc2 = gene2.loc[gene2['structure_abbreviation']=='VFC']
dfc2 = gene2.loc[gene2['structure_abbreviation']=='DFC']
mfc2 = gene2.loc[gene2['structure_abbreviation']=='MFC']
```

```python
#get means of gene1 expression level for each age
vfc2_mean = vfc2.groupby(['donor_age']).mean(['expression_level'])
dfc2_mean = dfc2.groupby(['donor_age']).mean(['expression_level'])
mfc2_mean = mfc2.groupby(['donor_age']).mean(['expression_level'])

fig, ax = plt.subplots(3)
x = ['0.3','1.0','2.0','3.0','4.0','8.0','11.0','13.0','18.0','19.0']
x1 = ['0.3','0.8','1.0','2.0','3.0','4.0','8.0','11.0','13.0','18.0','19.0']
v = vfc2_mean['expression_level']
d = dfc2_mean['expression_level']
m = mfc2_mean['expression_level']


ax[0].bar(x,v,align='center',color='lavender',label='Gene 2 Expression, VFC')
ax[0].plot(x,v,color='black')

ax[1].bar(x1,d,align='center',color='darkseagreen',label='Gene 2 Expression, DFC')
ax[1].plot(x1,d,color='black')

ax[2].bar(x,m,align='center',color='powderblue',label='Gene 2 Expression, MFC')
ax[2].plot(x,m,color='black')

ax[0].legend()
ax[1].legend()
ax[2].legend()
```
